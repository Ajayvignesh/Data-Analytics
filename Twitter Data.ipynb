```python
df=pd.read_csv("twitterdata.csv")
df.columns=['time','tweets']
df = df.head(200)
df
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk
nltk.download('punkt')
```

```python
#tokenization and stemming
ps = PorterStemmer()
words_stem =[]
column=df['tweets']
for line in column:
  words = word_tokenize(line)
  for w in words:
    words_stem.append(ps.stem(w))

words_stem[:50]
```

```python
import re
words = []
for w in words_stem:
  match = re.search(r'[A-Za-z]+',w)
  if match:
    words.append(w)
```

```python
#n grams
from nltk import ngrams


threegrams = ngrams(words, 3)

for grams in threegrams:
  print(grams)
  
```

```python
#document term matrix
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
X = vec.fit_transform(column)
df_term = pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
print(df_term)
```
